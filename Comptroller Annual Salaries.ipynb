# HYPOTHESIS 1

The mean of annual salaries for the Department of Corrections increases by 2% every year.

```python
#creating a dataframe which contains only department of Corrections rows.
df_doc = merge2[(merge2['Department Name'] == 'Department of Corrections')]
df_doc.head()
```

```python
mean_salary_by_year = df_doc.groupby('Year')['Annual Salary'].mean()
mean_salary_by_year
```

```python
mean_salary_percent_change=[0]
for a, b in zip(mean_salary_by_year[::1], mean_salary_by_year[1::1]):
    print(a,b)
    mean_salary_percent_change.append(100 * (b - a) / a)
mean_salary_percent_change
```

```python
plt.plot(mean_salary_percent_change)
plt.xlabel('Years')
plt.ylabel('The change in percentage')
plt.title('Percentage change of Mean Annual Salary for Department Of Corrections in years 2011-2015')
positions = (0, 1, 2, 3)
labels = ("2011", "2013", "2014", "2015")
plt.xticks(positions, labels)
plt.show()
```

My Hypothesis is False as the mean annual salary for Department of Corrections is not increased by 2% every year.

# QUESTION 1

What is the Name of the person who has the highest percent change in the annual salary when compared 2011 and 2015 annual salaries?

```python
#creating a dataframe which contains only 2011 and 2015 rows.
df_2011_and_2015 = merge2.loc[merge2['Year'].isin(['2011','2015'])]
df_2011_and_2015.head()
```

A dictionary is taken where the list of Names present were appended as keys and the Annual salaries of 2011 and 2015 as the values for the keys. The multiple key values are added through iterating the groupby dataframe and accessing them. The groupby Dataframe has a series datatype

```python
employee_salary=df_2011_and_2015.groupby(['Name','Year'])['Annual Salary'].first()
list_of_names=df_2011_and_2015['Name'].unique()
dict_name={}
for i in list_of_names:
    dict_name[i] = []
for i in employee_salary.iteritems():
    dict_name[i[0][0]].append(i[1])
dict_name
```

The percentage change is calculated by accessing the values from the dictionary. 
Some people who are in 2011 are not in 2015(reason they maybe retired) and  people who are in 2015 not in 2011(reason they maybe appointed newly). So, for these people exceptional math error is calculated.

```python
percent_change_in_salary=[]
for key,value in dict_name.items():
    try:
        percent_change_in_salary.append(100*(dict_name[key][-1]-dict_name[key][0])/dict_name[key][0])    
    except ZeroDivisionError:
        percent_change_in_salary.append(0)
```

```python
index_of_max_percent_change_in_salary = percent_change_in_salary.index(max(percent_change_in_salary))
print('The highest percentage change in Annual Salary from 2011 to 2015 is for ',list_of_names[index_of_max_percent_change_in_salary])
```

rounding the percent_change_in_salary to integers and storing in a list.

```python
newlist = [round(x) for x in percent_change_in_salary]  
```

The list with percent change in salary is convered to dataframe to plot histogram.

```python
to_plot=pd.DataFrame(newlist, columns=['Perecent Change'])
```

```python
to_plot.head()
```

```python
to_plot['Perecent Change'].value_counts().plot.hist(bins = 50, title='HIstogram of Annual Salary Difference')
plt.xlim(0, 1000)
plt.xlabel('Change in Percent')
plt.ylabel('Number of People')
plt.show()
```
